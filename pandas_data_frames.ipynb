## Your best friend: `groupby`
Problem is that if you have too many different values, it won't be visually useful

````python
df.groupby('neighbourhood').price.mean().plot.bar()
````

Instead, you would probably prefere to select specific values from your desired column first

````python
bush_harl_soslo_bool = df.neighbourhood.isin(['Bushwick', 'Harlem', 'South Slope'])
````

````python
bush_harl_soslo = df[bush_harl_soslo_bool]
bush_harl_soslo.head()
````

````python
bush_harl_soslo.groupby('neighbourhood').price.mean().sort_values(ascending=False).plot.bar()
````

`groupby` works very well with columns that have few unique values, such as 'neighbourhood_group'

````python
df.groupby('neighbourhood_group').price.mean().plot.bar()
````

## Creating column based on conditions

We are going to use `np.select`, which needs a list of conditions and a list of values to apply for each condition:

````python
import numpy as np
conditions = [
    (df['price'] <= 50),
    (df['price'] > 50) & (df['price'] <= 100),
    (df['price'] > 100)
    ]

# create a list of the values we want to assign for each condition
values = ['low', 'medium', 'high']

# create a new column and use np.select to assign values to it using our lists as arguments
df['price_range'] = np.select(conditions, values)
````

````python
df.head()
````

````python
df.groupby('price_range').size().loc[['low', 'medium', 'high']].plot.bar()
````

````python
df.groupby(['neighbourhood_group', 'price_range']).size().plot.bar()
````

````python
df2 = df[(df.room_type == 'Entire home/apt') & (df.price >= 30) & (df.price <=500)]
````

````python
harl = df2[df2.neighbourhood == "Harlem"]
bush = df2[df2.neighbourhood == "Bushwick"]
soslo = df2[df2.neighbourhood == "South Slope"]
print(harl.price.mean())
print(bush.price.mean())
print(soslo.price.mean())
````

````python
habu = pd.concat([harl, bush, soslo])
habu.neighbourhood.value_counts()
````

````python
plot = habu.groupby('neighbourhood').price.mean().plot.barh(
    title='My great chart', 
    color="red",
    xlim=(30,200),
    ylabel='price',
    xlabel='hoods')
fig = plot.get_figure()
fig.savefig('crazy_graph.jpg',dpi=300, bbox_inches = "tight")
````
